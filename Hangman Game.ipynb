```python
import random

print("âœ… Hangman Game Initialized!")
print("ðŸŽ¯ We'll use: random, while loops, if-else, strings, lists")
```

Define Game Variables and Word List

```python
# Predefined list of words (simplified scope - 5 words)
words = ["python", "hangman", "computer", "program", "coding"]

# Game variables
max_attempts = 6  # Limited to 6 incorrect guesses

print("ðŸ“ Word List Ready:")
for i, word in enumerate(words, 1):
    print(f"  {i}. {word} (length: {len(word)})")

print(f"\ðŸŽ¯ Game Rules:")
print(f"  â€¢ Maximum incorrect guesses: {max_attempts}")
print(f"  â€¢ Guess one letter at a time")
print(f"  â€¢ Words are related to programming")
```

 Create Display Function

```python
def display_game_state(word, guessed_letters, attempts_left, incorrect_guesses):
    """
    Display the current state of the game
    """
    print("\n" + "=" * 40)
    
    # Display the word with blanks and guessed letters
    display_word = ""
    for letter in word:
        if letter in guessed_letters:
            display_word += letter + " "
        else:
            display_word += "_ "
    
    print(f"Word: {display_word}")
    print(f"Attempts left: {attempts_left}")
    print(f"Incorrect guesses: {', '.join(incorrect_guesses) if incorrect_guesses else 'None'}")
    print(f"Guessed letters: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
    
    # Simple hangman visualization
    hangman_stages = [
        """
          -----
          |   |
          |
          |
          |
          |
        ========
        """,
        """
          -----
          |   |
          |   O
          |
          |
          |
        ========
        """,
        """
          -----
          |   |
          |   O
          |   |
          |
          |
        ========
        """,
        """
          -----
          |   |
          |   O
          |  /|
          |
          |
        ========
        """,
        """
          -----
          |   |
          |   O
          |  /|\\
          |
          |
        ========
        """,
        """
          -----
          |   |
          |   O
          |  /|\\
          |  /
          |
        ========
        """,
        """
          -----
          |   |
          |   O
          |  /|\\
          |  / \\
          |
        ========
        """
    ]
    
    wrong_count = len(incorrect_guesses)
    print(hangman_stages[wrong_count])

print("âœ… Display function created!")
```

 Create Main Game Logic

```python
def play_hangman():
    """
    Main function to play the Hangman game
    """
    # Select a random word
    secret_word = random.choice(words)
    guessed_letters = set()
    incorrect_guesses = []
    attempts_left = max_attempts
    
    print("ðŸŽ® WELCOME TO HANGMAN!")
    print("=" * 50)
    print("Guess the programming-related word!")
    print("Type 'quit' to exit the game")
    print("=" * 50)
    
    # Main game loop
    while attempts_left > 0:
        # Display current game state
        display_game_state(secret_word, guessed_letters, attempts_left, incorrect_guesses)
        
        # Check if player won
        if all(letter in guessed_letters for letter in secret_word):
            print(f"ðŸŽ‰ CONGRATULATIONS! You guessed the word: {secret_word.upper()}")
            print("ðŸ† YOU WIN!")
            break
        
        # Get player's guess
        guess = input("Enter a letter: ").lower().strip()
        
        # Check for quit command
        if guess == 'quit':
            print(f"ðŸ‘‹ Thanks for playing! The word was: {secret_word}")
            return
        
        # Validate input
        if len(guess) != 1:
            print("âŒ Please enter exactly one letter!")
            continue
            
        if not guess.isalpha():
            print("âŒ Please enter a valid letter (a-z)!")
            continue
            
        if guess in guessed_letters:
            print("âŒ You already guessed that letter!")
            continue
        
        # Add to guessed letters
        guessed_letters.add(guess)
        
        # Check if guess is correct
        if guess in secret_word:
            print(f"âœ… Good guess! '{guess}' is in the word!")
        else:
            print(f"âŒ Sorry, '{guess}' is not in the word!")
            incorrect_guesses.append(guess)
            attempts_left -= 1
    
    # Game over - player lost
    if attempts_left == 0:
        display_game_state(secret_word, guessed_letters, attempts_left, incorrect_guesses)
        print(f"ðŸ’€ GAME OVER! The word was: {secret_word.upper()}")
        print("Better luck next time!")

print("âœ… Main game logic created!")
print("ðŸš€ Ready to play Hangman!")
```

Enhanced Version with Score Tracking

```python
def enhanced_hangman():
    """
    Enhanced version with score tracking and multiple rounds
    """
    wins = 0
    losses = 0
    
    while True:
        # Play one round
        secret_word = random.choice(words)
        guessed_letters = set()
        incorrect_guesses = []
        attempts_left = max_attempts
        
        print(f"\nðŸŽ® ROUND STATS - Wins: {wins} | Losses: {losses}")
        print("=" * 50)
        
        # Round loop
        while attempts_left > 0:
            display_game_state(secret_word, guessed_letters, attempts_left, incorrect_guesses)
            
            # Check win condition
            if all(letter in guessed_letters for letter in secret_word):
                print(f"ðŸŽ‰ YOU WIN! The word was: {secret_word.upper()}")
                wins += 1
                break
            
            # Get guess
            guess = input("Enter a letter (or 'quit' to exit): ").lower().strip()
            
            if guess == 'quit':
                print(f"ðŸ‘‹ Final Score - Wins: {wins}, Losses: {losses}")
                return
            
            # Validate guess
            if len(guess) != 1 or not guess.isalpha():
                print("âŒ Please enter a single letter (a-z)!")
                continue
                
            if guess in guessed_letters:
                print("âŒ Already guessed that letter!")
                continue
            
            guessed_letters.add(guess)
            
            if guess in secret_word:
                print(f"âœ… Correct! '{guess}' is in the word!")
            else:
                print(f"âŒ Wrong! '{guess}' is not in the word!")
                incorrect_guesses.append(guess)
                attempts_left -= 1
        
        # Handle loss
        if attempts_left == 0:
            display_game_state(secret_word, guessed_letters, attempts_left, incorrect_guesses)
            print(f"ðŸ’€ You lost! The word was: {secret_word.upper()}")
            losses += 1
        
        # Ask to play again
        play_again = input("\nPlay again? (y/n): ").lower().strip()
        if play_again != 'y':
            print(f"ðŸ‘‹ Final Score - Wins: {wins}, Losses: {losses}")
            print("Thanks for playing!")
            break

print("âœ… Enhanced Hangman ready!")
print("ðŸ† Now with score tracking and multiple rounds!")
```
