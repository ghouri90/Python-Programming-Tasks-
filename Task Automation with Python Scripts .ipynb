Create JPG Moving Function

```python
def move_jpg_files_safe(source_path, dest_path):
    """
    Safely move all JPG files from source to destination
    """
    print("üöÄ Starting JPG file organization...")
    print("-" * 50)
    
    # Create destination folder
    Path(dest_path).mkdir(parents=True, exist_ok=True)
    print(f"üìÅ Destination folder ready: {dest_path}")
    
    # Check if source exists
    if not os.path.exists(source_path):
        return {'success': False, 'message': 'Source folder does not exist'}
    
    # Get JPG files
    all_files = os.listdir(source_path)
    jpg_files = [f for f in all_files if f.lower().endswith('.jpg')]
    
    print(f"üéØ Found {len(jpg_files)} JPG files to process")
    
    if not jpg_files:
        return {'success': False, 'message': 'No JPG files found'}
    
    # Move files
    moved_files = []
    error_files = []
    
    for jpg_file in jpg_files:
        try:
            source_file = os.path.join(source_path, jpg_file)
            dest_file = os.path.join(dest_path, jpg_file)
            
            # Check if file exists
            if not os.path.exists(source_file):
                error_files.append((jpg_file, "File not found"))
                continue
            
            # Handle duplicates
            counter = 1
            final_dest = dest_file
            while os.path.exists(final_dest):
                name, ext = os.path.splitext(jpg_file)
                final_dest = os.path.join(dest_path, f"{name}_{counter}{ext}")
                counter += 1
            
            # Move the file
            shutil.move(source_file, final_dest)
            moved_files.append(jpg_file)
            print(f"‚úÖ Moved: {jpg_file}")
            
        except Exception as e:
            error_files.append((jpg_file, str(e)))
            print(f"‚ùå Error: {jpg_file} - {str(e)}")
    
    return {
        'success': True,
        'moved_count': len(moved_files),
        'error_count': len(error_files),
        'total_found': len(jpg_files),
        'moved_files': moved_files,
        'error_files': error_files
    }

print("‚úÖ JPG moving function created successfully!")
```

 Execute JPG File Organization

```python
# Execute JPG file moving
print("üîÑ EXECUTING JPG FILE ORGANIZATION")
print("=" * 50)

start_time = time.time()
jpg_result = move_jpg_files_safe(source_folder, destination_folder)
end_time = time.time()
execution_time = end_time - start_time

print("-" * 50)
print("üìä JPG ORGANIZATION COMPLETED!")
```

Create Email Extraction Function

```python
def extract_emails_from_file(input_file, output_file):
    """
    Extract all email addresses from a text file
    """
    print("üîç Starting email extraction...")
    print("-" * 50)
    
    # Check if input file exists
    if not os.path.exists(input_file):
        return {'success': False, 'message': 'Input file does not exist'}
    
    try:
        # Read the file
        with open(input_file, 'r', encoding='utf-8') as file:
            content = file.read()
        
        print(f"üìñ Read {len(content)} characters from file")
        
        # Email pattern
        email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
        
        # Find all emails
        emails_found = re.findall(email_pattern, content)
        
        # Remove duplicates
        unique_emails = []
        for email in emails_found:
            if email not in unique_emails:
                unique_emails.append(email)
        
        print(f"üéØ Found {len(emails_found)} email instances")
        print(f"üìß {len(unique_emails)} unique emails")
        
        # Save to output file
        if unique_emails:
            with open(output_file, 'w', encoding='utf-8') as file:
                file.write("EXTRACTED EMAIL ADDRESSES\n")
                file.write("=" * 40 + "\n")
                file.write(f"Source: {input_file}\n")
                file.write(f"Date: {time.strftime('%Y-%m-%d %H:%M:%S')}\n")
                file.write(f"Total: {len(unique_emails)} unique emails\n")
                file.write("=" * 40 + "\n\n")
                
                for i, email in enumerate(unique_emails, 1):
                    file.write(f"{i:2d}. {email}\n")
            
            print(f"üíæ Saved to: {output_file}")
        
        return {
            'success': True,
            'emails_found': len(emails_found),
            'unique_emails': len(unique_emails),
            'emails_list': unique_emails
        }
        
    except Exception as e:
        return {'success': False, 'message': f'Error: {str(e)}'}

print("‚úÖ Email extraction function created!")
```

Execute Email Extraction

```python
# Execute email extraction
print("üîÑ EXECUTING EMAIL EXTRACTION")
print("=" * 50)

start_time_email = time.time()
email_result = extract_emails_from_file(input_text_file, output_emails_file)
end_time_email = time.time()
email_time = end_time_email - start_time_email

print("-" * 50)
print("üìß EMAIL EXTRACTION COMPLETED!")
```
